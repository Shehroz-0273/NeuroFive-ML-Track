```python
# Data Story:
'''The Titanic dataset contains information about passengers aboard the Titanic
and whether they survived the disaster. 
It consists of 891 rows and 12 columns with both numerical and categorical
features. During the initial exploration, I found that some columns, such as Age,
Cabin, and Embarked, contain missing values.
The dataset includes important information like passenger class, age, gender, 
fare, and survival status. 
This exploratory data analysis helped me understand the structure and quality
of the dataset before applying any machine learning techniques.'''
```

```python
#  Week-1 (Part-2)
```

```python
# Fill Ages with Median
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
# Cabin Drop
df = df.drop(columns = ["Cabin"])
```

```python
# Delete Rows for the missing values of Embarked
df = df.dropna(subset = ["Embarked"])
```

```python
# Handling Missing Values
'''The Age column had 177 missing values, so I filled them using the median
because age may contain outliers and the median is more robust than the mean.
The Cabin column had 687 missing values, so I removed it because of its data was
missing.
The Embarked column had only 2 missing values, so I removed those rows since they
would have minimal impact on the dataset.'''
```

```python
# Visualization Libraries of Python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (8,5))
sns.boxplot(x = df["Fare"])
plt.title("Boxplot of Fare")
plt.show()
```

```python
# Outlier Detection
'''
I used a boxplot to detect outliers in the Fare column. The visualization show 
several values that are much higher than the rest of the data, indicating the 
presence of outliers. These high fares may represent first-class passengers or 
special ticket prices.
'''
```

```python
plt.figure(figsize = (8,5))
sns.histplot(df["Age"], bins = 20)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar Chart
plt.figure(figsize=(6,4))

sns.countplot(x="Sex", hue="Survived", data=df)

plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.legend(title="Survived", labels=["No", "Yes"])

plt.show()
```

```python
# Correlation HeatMap
plt.figure(figsize=(10,6))

correlation = df.corr(numeric_only=True)

sns.heatmap(correlation,
            annot=True,
            cmap="coolwarm",
            fmt=".2f")

plt.title("Correlation Heatmap")
plt.show()
```
